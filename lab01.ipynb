```python
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
img = cv2.imread('/content/dog.jpg')
```

```python
plt.imshow(img)
plt.show()
```

```python
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(img_rgb)
plt.show()
```

```python
img_rgb.shape
```

```python
type(img_rgb)
```

```python
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

```python
plt.imshow(gray)
plt.show()
```

```python
plt.imshow(gray, cmap= 'gray')
plt.show()
```

```python
gray.shape
```

```python
type(gray)
```

```python
plt.imshow(img_rgb[:,:,0],cmap= 'gray')
plt.show()
```

```python
plt.imshow(img_rgb[:,:,1],cmap= 'gray')
plt.show()
```

```python
plt.imshow(img_rgb[:,:,2],cmap= 'gray')
plt.show()
```

```python
#Create a figure
fig = plt.figure(figsize=(8,6)) #Specify figure size in inches
# Add a single subplot to the figure
ax = fig.add_subplot(111)  # (1 row,1 column,1st subplot)
# Display the image
ax.imshow(img_rgb)
ax.axis('off') # Turnoff axis and labels and ticks
ax.set_title("Image display") # Set title for the subplot
# Show the figure
plt.show()
```

```python
img_rgb.shape
```

```python
resized = cv2.resize(img_rgb, (300,300))
```

```python
resized.shape
```

```python
fig = plt.figure(figsize=(8,6))
ax = fig.add_subplot(111)
ax.imshow(resized)
ax.axis('off')
ax.set_title("Image Display")
plt.show()
```

```python
cv2.line(img_rgb, (50,50), (200,50), (255,0,0), 5)
cv2.rectangle(img_rgb, (60,60), (160,160), (0,255,0), 3)
cv2.circle(img_rgb, (120,120), 40, (0,0,255), -1)
plt.imshow(img_rgb)
plt.show()
```
